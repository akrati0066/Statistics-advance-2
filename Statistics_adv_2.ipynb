```python
#Question 1: What is hypothesis testing in statistics?
#Hypothesis testing is a statistical method to check assumptions about a population using sample data. We set a null hypothesis (H₀) and an alternative hypothesis (H₁), calculate a test statistic, and decide (using p-value or critical value) whether to reject H₀ or not.
```

```python
#Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
#Null hypothesis (H₀): It is the assumption that there is no effect or no difference (the status quo).
#Alternative hypothesis (H₁/Ha): It states that there is an effect or a difference.
#Example:
#H₀: A medicine has no effect.
#H₁: The medicine has an effect.
#Difference: H₀ assumes no change, while H₁ assumes a change exists.
```

```python
#Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
#Significance level (α): It is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01.
#Role: It sets the cutoff for decision-making. If the p-value ≤ α, we reject H₀; otherwise, we fail to reject H₀
```

```python
#Question 4: What are Type I and Type II errors? Give examples of each.
#Type I Error (α): Rejecting the null hypothesis when it is actually true (false positive).
#Example: Concluding a medicine works when it actually doesn’t.
#Type II Error (β): Failing to reject the null hypothesis when it is false (false negative).
#Example: Concluding a medicine doesn’t work when it actually does.
```

```python
#Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
#Z-test: Used when the population variance is known and the sample size is large (n > 30).
#T-test: Used when the population variance is unknown and the sample size is small (n ≤ 30).
#Example:
#Z-test: Testing average weight of 100 students when population variance is known.
#T-test: Testing average marks of 20 students when variance is unknown.
```

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis parameters
mu_0 = 50   # population mean under null hypothesis
alpha = 0.05  # significance level

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Print results
print("Sample mean:", sample_mean)
print("Sample std deviation:", sample_std)
print("Sample size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # true mean and standard deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Step 3: 95% confidence interval (using t-distribution)
confidence = 0.95
ci = stats.t.interval(confidence, df=n-1, loc=sample_mean, scale=se)

# Print results
print("Sample mean:", sample_mean)
print("95% Confidence Interval:", ci)

# Step 4: Plot histogram
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci[0], color='green', linestyle='--', label=f"Lower CI = {ci[0]:.2f}")
plt.axvline(ci[1], color='green', linestyle='--', label=f"Upper CI = {ci[1]:.2f}")
plt.title("Normal Distribution with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    # Calculate mean and std deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)

    # Compute Z-scores
    z_scores = (data - mean) / std

    # Plot histogram
    plt.hist(z_scores, bins=15, color='lightblue', edgecolor='black')
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(50, 10, 100)  # simulate dataset
z_scores = calculate_z_scores(data)
print("First 10 Z-scores:", z_scores[:10])
```
